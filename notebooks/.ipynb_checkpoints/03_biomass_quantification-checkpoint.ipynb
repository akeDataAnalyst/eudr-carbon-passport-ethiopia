#### Loading Audit Results & Carbon Coefficients

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os

# Load spatial results
audit_path = '../data/processed/spatial_audit_results.geojson'
try:
    farms = gpd.read_file(audit_path, engine='fiona')
except:
    import json
    with open(audit_path, 'r') as f:
        data = json.load(f)
    farms = gpd.GeoDataFrame.from_features(data['features'], crs="EPSG:4326")

# Load NDVI time-series
ndvi_df = pd.read_csv('../data/processed/ndvi_baseline_2019_2025.csv')

AGB_COEFF = 120
CARBON_FRACTION = 0.47
CO2_CONVERSION = 3.67

print("Ready for Carbon Modeling.")
```

#### Linking Vegetation Health to Biomass

```python
# Calculate the 'Vegetation Health Index' for the most recent year (2025)
recent_ndvi = ndvi_df[ndvi_df['observation_date'].str.contains('2025')]['ndvi_value'].mean()

# Normalized Biomass Factor (assuming NDVI of 0.8 is 100% biomass capacity)
biomass_health_factor = np.clip(recent_ndvi / 0.8, 0, 1)

# Calculate Tonnes of CO2 equivalent per hectare
# Formula: (Base Biomass * Health Factor) * Carbon % * CO2 Ratio
carbon_per_ha = (AGB_COEFF * biomass_health_factor) * CARBON_FRACTION * CO2_CONVERSION

print(f"Mean NDVI (2025): {recent_ndvi:.3f}")
print(f"Biomass Health Factor: {biomass_health_factor:.2%}")
print(f"Estimated Sequestration: {carbon_per_ha:.2f} tCO2e / Hectare")
```

#### Calculating Total Farm Assets

```python
# Apply to the farm dataframe
farms['carbon_per_ha'] = carbon_per_ha
farms['total_carbon_stock'] = farms['carbon_per_ha'] * farms['area_ha']

# Current price per voluntary carbon credit (tCO2e) is roughly $15 - $30
CARBON_PRICE_USD = 25.0 
farms['carbon_asset_value_usd'] = farms['total_carbon_stock'] * CARBON_PRICE_USD

# Display the Financial Narrative
total_stock = farms['total_carbon_stock'].iloc[0]
total_value = farms['carbon_asset_value_usd'].iloc[0]

print(f"CARBON LEDGER: {farms['farm_id'].iloc[0]}")
print(f"Total Carbon Stock: {total_stock:.2f} Tonnes CO2e")
print(f"Estimated Market Value: ${total_value:,.2f} USD")
```

#### Saving the Carbon Passport

```python
# Save the final consolidated audit results
final_output = '../data/processed/final_carbon_passport.geojson'

if os.path.exists(final_output):
    os.remove(final_output)

try:
    farms.to_file(final_output, driver='GeoJSON', engine='fiona')
    print("Carbon Passport saved successfully.")
except:
    with open(final_output, 'w') as f:
        f.write(farms.to_json())
    print("Carbon Passport saved (JSON Fallback).")
```
